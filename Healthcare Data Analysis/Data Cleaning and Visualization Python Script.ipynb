# Coding Task (Q7)

```python
def unique_em(emails):
    unique_emails = set()

    for email in emails:
        local, domain = email.split('@')
        local = local.split('+')[0].replace('.', '')  # Applying the transformations
        clean_email = f"{local}@{domain}"
        unique_emails.add(clean_email)

    return len(unique_emails)

list1 = ["hello.world+april@example.com", "april.2024.spring+lovely.weather@example.com", "helloworld@example.com"]
list2 = ["a@utahdhhs.com", "b@utahdhhs.com", "c@utahdhhs.com"]

print("Unique number of Email Addresses from list 1 - ",unique_em(list1))
print("Unique number of Email Addresses from list 2 - ",unique_em(list2))
```

# Data Cleaning Automation Pipeline

### Patients Admission

```python
import pandas as pd

file_path = 'PatientAdmission.txt'
data = pd.read_csv(file_path, delimiter=',', header=None, names=['ID', 'Name', 'Age', 'Admission_Date'])

def validate_id(id):
    return id.isdigit()

def validate_name(name):
    return name.isalpha()

def validate_age(age):
    return age.isdigit() and 0 <= int(age) <= 120

def validate_date(date):
    try:
        pd.to_datetime(date, format='%YYYY-%MM-%dd')
        return True
    except ValueError:
        return False

data['ID_Valid'] = data['ID'].apply(validate_id)
data['Name_Valid'] = data['Name'].apply(validate_name)
data['Age_Valid'] = data['Age'].apply(validate_age)
data['Date_Valid'] = data['Admission_Date'].apply(validate_date)

valid_data = data[(data['ID_Valid'] & data['Name_Valid'] & data['Age_Valid'] & data['Date_Valid'])]

valid_data.to_csv('ValidatedPatientAdmission.csv', index=False)

print("Data validation complete and saved to 'ValidatedPatientAdmission.csv'.")
```

# Future Scope

> Code based ETL pipeline

* An automated, code-driven ETL pipeline can be developed to efficiently handle data workflows. This pipeline would fetch raw data from various sources, such as databases or spreadsheets, perform necessary cleaning and transformations, and then load the processed data back into a seperate table in database

* This process can be automated using a tool like Apache Airflow, which would schedule and run the pipeline at regular intervals. The system would append the transformed data to our database, ensuring up-to-date information is always available. This data can then be linked to visualization tools (Power BI/ Tableau) for enhanced data analysis and reporting

* Continuous improvements can be made to the pipeline, particularly in enhancing its robustness through better exception handling and error logging. This iterative development can lead to a more resilient and efficient ETL process over time.

> Cloud based ETL pipeline

* As data volumes grow, the limitations of on-premises databases become apparent. A cloud-based ETL pipeline offers a scalable and efficient alternative. Utilizing cloud services from providers like AWS or Azure facilitates handling larger datasets and allows for more complex processing without the hardware limitations of on-prem systems

* Cloud-based solutions also offer advantages in terms of scalability, reliability, and security. They can dynamically allocate resources based on the pipeline's load, perform real-time data processing, and ensure data integrity with advanced security measures. Transitioning to a cloud-based ETL can reduce operational overhead and improve data handling capabilities.

* This approach not only future-proofs the data management strategy but also integrates well with other cloud-based analytics and machine learning services, enabling more advanced data insights and decision-making capabilities
